```python
import pandas as pd

# Cargar el CSV
df_features = pd.read_csv('caracteristicas_seleccionadas.csv', header=None)
df_features.columns = ['ETF', 'Date','Method'] + [f'Feature_{i}' for i in range(10)]
df_features = df_features[df_features['Method'] != 'estatico']
print(df_features.head())
```

```python
df_features['Date'] = pd.to_datetime(df_features['Date'])

# Unimos todas las columnas de características en una sola columna
features_columns = [f'Feature_{i}' for i in range(10)]
df_melted = df_features.melt(id_vars=['ETF', 'Date', 'Method'], value_vars=features_columns, var_name='Feature_Num', value_name='Feature')

# Mostramos las primeras filas del dataframe transformado
print(df_melted.tail())
```

```python
# Contar la frecuencia de cada característica
feature_counts = df_melted['Feature'].value_counts()
print(feature_counts.head(10))

# Guardar o mostrar el conteo en un archivo
feature_counts.to_csv('feature_counts.csv')
```

```python
feature_counts = feature_counts.head(20)
caracteristicas = list(feature_counts.index)
caracteristicas
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def analizar_parametros():
    # Leer el archivo CSV
    df = pd.read_csv('hiperparametros_seleccionados.csv')
    
    # Análisis descriptivo básico
    descripcion = df.describe()
    
    # Frecuencias individuales de los parámetros
    frecuencias = {col: df[col].value_counts() for col in df.columns if col != 'ETF' and col != 'date'}
    
    # Visualización de distribuciones para cada parámetro numérico
    fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
    fig.subplots_adjust(hspace=0.5, wspace=0.5)
    axs = axs.flatten()
    
    for i, col in enumerate(['subsample', 'reg_lambda', 'reg_alpha', 'n_estimators', 'max_depth', 'learning_rate', 'gamma', 'colsample_bytree']):
        df[col].hist(ax=axs[i])
        axs[i].set_title(f'Distribución de {col}')
        axs[i].set_xlabel(col)
        axs[i].set_ylabel('Frecuencia')
    
    plt.show()
    
    return descripcion, frecuencias

descripcion, frecuencias = analizar_parametros()
#print(descripcion)
for param, freq in frecuencias.items():
    print(f'Frecuencia de {param}:\n{freq}\n')
```
