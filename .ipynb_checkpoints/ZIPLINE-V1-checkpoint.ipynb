```python
ticker = yf.Ticker("AAPL")
data = ticker.history(start="2020-01-01", end="2020-12-31")
```

```python
data.index = data.index.tz_localize('UTC')
data = data.reset_index()

# Renombrar las columnas a minúsculas, ya que yfinance usa mayúsculas
data.columns = data.columns.str.lower()

data.rename(columns={'stock splits': 'split'}, inplace=True)
data.rename(columns={'dividends': 'dividend'}, inplace=True)
data['date'] = pd.to_datetime(data['date'], utc=True)

# Asegurarse de que el DataFrame contenga las columnas esperadas por Zipline
print(type(data))
#data.head()
data.tail()
#date,open,high,low,close,volume,dividend,split
```

```python
from zipline.api import order_target, record, symbol
from zipline import run_algorithm
from datetime import datetime
import pytz

def initialize(context):
    context.asset = symbol('AAPL')
    context.set_benchmark(symbol('AAPL'))  # Establece AAPL como el benchmark
    context.has_ordered = False  # Para controlar las operaciones mensuales

def handle_data(context, data):
    # Obtiene el día actual
    today = data.current_session

    # Chequea si es el primer día del mes
    if today.day == "01" and not context.has_ordered:
        # Ordena comprar una acción de AAPL
        order_target(context.asset, 1)
        context.has_ordered = True
    elif today.day != "01":
        context.has_ordered = False  # Resetea el indicador para el próximo mes

    # Registra el precio de cierre diario de AAPL
    record(AAPL=data.current(context.asset, 'price'))

# Define las fechas de inicio y fin del backtest
start = pd.Timestamp('2020-01-02', tz = pytz.UTC)
end = pd.Timestamp('2020-12-30', tz = pytz.UTC)

#start = datetime(2020, 1, 2, tzinfo=pytz.UTC)
#end = datetime(2020, 12, 30, tzinfo=pytz.UTC)

# Ejecuta el backtest
results = run_algorithm(
    start=start, end=end,
    initialize=initialize,
    handle_data=handle_data,
    capital_base=10000,
    data_frequency='daily',  # Aunque operamos mensualmente, los datos son diarios
    bundle='my_custom_bundle'  # Usa el nombre de tu paquete de datos personalizado
)
```
