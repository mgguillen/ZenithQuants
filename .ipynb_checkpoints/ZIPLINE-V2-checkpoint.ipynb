```python
from zipline.api import order_target, record, symbol, order_target_percent
from zipline import run_algorithm
from datetime import datetime
import random
import pytz


def initialize(context):
# Crea un activo para cada símbolo y almacénalos en una lista
    context.assets = [symbol(s) for s in ["XLE", "XLB", "XLI", "XLK", "XLF", "XLP", "XLY", "XLV", "XLU", "IYR", "VOX"]]   
    # Establece SPY como el benchmark
    context.set_benchmark(symbol('SPY'))    
    context.has_ordered = False
    
def handle_data(context, data):
    # Obtiene el día actual
    today = data.current_session
    
    num_assets = len(context.assets)
    #print(num_assets)
    

    valor_aleatorio = random.random()
    #print(today.day)
    
    # Chequea si es el primer día del mes
    if today.day == 1:
        # Si es el primer día del mes, compra un porcentaje fijo del capital disponible para cada activo
        for asset in context.assets:
            #print(asset)
            if data.can_trade(asset):
                print("comprar:",data.current(asset, 'price'))
                #order_target_percent(asset, 1 / num_assets)
                order_target(asset, 10 )
    
    # Si es el último día del mes, vende todo
    elif today.day == today.days_in_month:
        for asset in context.assets:
            if data.can_trade(asset):
                print("vender:",data.current(asset, 'price'))
                order_target_percent(asset, 0)
            else:
            # En cualquier otro día del mes, no hagas nada
                pass
    
    # Registra el valor de la cartera y el precio de cierre diario para cada activo
    record(Valor_Cartera=context.portfolio.portfolio_value)
    for asset in context.assets:
        current_price = data.current(asset, 'price')
        #print(current_price)
        # Usa el símbolo del activo como clave para registrar su precio actual
        record(**{asset.symbol: current_price})



# Define las fechas de inicio y fin del backtest
start = pd.Timestamp('2020-01-02', tz = pytz.UTC)
end = pd.Timestamp('2024-04-10', tz = pytz.UTC)

#start = datetime(2020, 1, 2, tzinfo=pytz.UTC)
#end = datetime(2020, 12, 30, tzinfo=pytz.UTC)

# Ejecuta el backtest
results = run_algorithm(
    start=start, end=end,
    initialize=initialize,
    handle_data=handle_data,
    capital_base=100000000000,
    data_frequency='daily',  # Aunque operamos mensualmente, los datos son diarios
    bundle='my_custom_bundle'  # Usa el nombre de tu paquete de datos personalizado
)
```
